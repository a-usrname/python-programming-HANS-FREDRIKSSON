##### (a) First, read the college data set into Python, by using the pd.read_csv() function.

```python
import pandas as pd

college = pd.read_csv('College.csv')

```

##### (b) The first column is Unnamed: 0 in the college data set and that column contains the name of each of the 777 colleges . 

```python
college

```

##### The college names are not needed for the data processing, thus the column is set to be the index instead (one college name per row). The column named Private will be the first column instead.

```python
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')

college = college3
college

```

##### c) Using the describe method to produce a numerical summary of the variables in the data set.

```python
numerical_summary = college.describe()
print(numerical_summary)

```

##### (d) Code for using pd.plotting.scatter_matrix function to produce a scatterplot matrix of the first columns (Top10perc, Apps & Enroll).

```python
# Source from stackoverflow.com, changed the column names 
# https://stackoverflow.com/questions/44884610/tplotting-scatter-matrix-with-selected-columns-from-panda-dataframe-using-scatte
pd.plotting.scatter_matrix(college[['Top10perc','Apps','Enroll']])

```

##### (e) Using the boxplot method of college to produce side-by-side boxplots of Outstate vs Private. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Outstate (numerical data) vs Private (categorical data) using Seaborn boxplot
# Source: https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot
outstate_vs_private = college[['Outstate', 'Private']]
sns.boxplot(data = outstate_vs_private, x='Private', y='Outstate')
# matplotlib to put name on the plot
plt.title('Distribution of Outstate vs Private')
plt.show()

```

##### (f) A new qualitative (yes/no) variable called Elite is created. Then used the value_counts() method of college to see how many elite colleges (78) there are, and then used the boxplot to produce side-by-side boxplots of Outstate vs Elite. 

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,50,100], labels=['No', 'Yes'])
# checking how many elite universities there are in the data set, by value_counts() method
print(college['Elite'].value_counts())

# boxplot side-by-side of Outstate vs Elite
outstate_vs_elite = college[['Outstate', 'Elite']]
sns.boxplot(data = outstate_vs_elite, x='Elite', y='Outstate')

plt.title('Distribution of Outstate vs Elite')
plt.show()

```

##### (g) Produced some histograms with differing numbers of bins for a few of the quantitative variables. I used 6 subplots in a 3x2 format.

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 8))

college['Apps'].plot.hist(bins=50, ax=axes[0,0], title='Number of applications received (50 bins)')
college['Accept'].plot.hist(bins=50, ax=axes[0,1], title='Number of applicants accepted (50 bins)')
college['Enroll'].plot.hist(bins=100, ax=axes[1,0], title='Number of new students enrolled (100 bins)')
college['Grad.Rate'].plot.hist(bins=100, ax=axes[1,1], title='Graduation Rate (100 bins)')
college['Books'].plot.hist(bins=150, ax=axes[2,0], title='Estimated book costs (150 bins)')
college['Personal'].plot.hist(bins=150, ax=axes[2,1], title='Estimated personal spending (150 bins)')

plt.tight_layout()
plt.show()

```

##### Boxplot of Terminal vs Elite: 
##### To analyse if colleges with Elite status tend to hire more faculty with higher academic degree(Terminal degrees).
##### The boxplot shows that colleges with Elite status do tend to hire more faculty with Terminal degrees, compared to colleges without Elite status.

```python
# terminal vs elite, boxplot
terminal_vs_elite = college[['Terminal', 'Elite']]

sns.boxplot(data = terminal_vs_elite, x='Elite', y='Terminal')

plt.title('Distribution of Terminal vs Elite')
plt.show()

```

##### Personal spending, instructional expenditure vs elite status.
##### Analyse correlations between personal spending, instructional expenditure and Elite status of the colleges.
##### It seems to be higher instructional expenditure at colleges with Elite status compared to colleges without Elite status. Personal spending seems to be a little bit higher at colleges without Elite status. 

```python
# Personal spending; Instructional expenditure, vs Elite status

# side-by-side box.plot for personal spending/elite status and instructional expenditure/elite:
personal_vs_elite = college[['Personal', 'Elite']]
expend_vs_elite = college[['Expend', 'Elite']]

fig, axes = plt.subplots(1, 2, figsize=(8, 10))
sns.boxplot(data = personal_vs_elite, ax=axes[0], x='Elite', y='Personal') 
# axes set_title(ChatGPT)
axes[0].set_title('Estimated personal spending vs Elite')
sns.boxplot(data = expend_vs_elite, ax=axes[1], x='Elite', y='Expend')
axes[1].set_title('Instructional expenditure per student vs Elite')

plt.tight_layout() #annars overlap
plt.show()  

```

##### Personal spending vs out of state tuition
##### To check for correlation. The correlation matrix shows a moderate negative correlation. This could mean that when personal spending increases, the out-of-state tuition tends to decrease. But, this correlation is not strong.


```python
# Personal spending, Out-of-state tuition (variables: Personal, Outstate)
# sum personal, outstate 
print('\nTotal:\n', college[['Personal', 'Outstate']].sum())

# corr matrix
correlation_matrix_per_out = college[['Personal', 'Outstate']].corr()
print(f'\nCorrelation matrix:\n {correlation_matrix_per_out}')

```
